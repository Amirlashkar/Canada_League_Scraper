```python
df= pd.DataFrame(filtered_rows1)

# Assuming the data is in a DataFrame named df
df['Time_sec'] = df['Minutes']*60 + df['seconds']

# Calculate the points scored and conceded when each player is on the court
df['points_scored'] = pd.to_numeric(df['Visitor score']).diff().where(df['First_True_Column'] == 'enters the game')
df['points_conceded'] = pd.to_numeric(df['Home score']).diff().where(df['First_True_Column'] == 'enters the game')

# Calculate the time each player spent on the court
df['enter_time'] = df['Time_sec'].where(df['First_True_Column'] == 'enters the game')
df['leave_time'] = df['Time_sec'].where(df['First_True_Column'] == 'goes to the bench')
df['time_on_court'] = df.groupby('Second_True_Column')['leave_time'].shift() - df['enter_time']

# Group by player and calculate total points scored, conceded and total time on court
player_stats = df.groupby('Second_True_Column').agg({'points_scored': 'sum', 'points_conceded': 'sum', 'time_on_court': 'sum'})

# Calculate net rating for each player
player_stats['net_rating'] = (player_stats['points_scored'] - player_stats['points_conceded']) / player_stats['time_on_court']

# Calculate net rating for the lineup
lineup = ['VREEKEN,CONNOR', 'WARNHOLTZ,AIDEN','SERAPH,REGINALD JEAN','LATIFF,WAZIR','SHEPHARD,GRANT']

# Check if all players in lineup are in player_stats
if set(lineup).issubset(player_stats.index):
    lineup_stats = player_stats.loc[lineup].sum()
    lineup_net_rating = (lineup_stats['points_scored'] - lineup_stats['points_conceded']) / lineup_stats['time_on_court']
    print('Lineup Net Rating:', lineup_net_rating)
else:
    missing_players = set(lineup) - set(player_stats.index)
    print(f'The following players in your lineup are not present in Second_True_Column: {missing_players}')

print(player_stats)
```

```python
starter = ['VREEKEN,CONNOR', 'WARNHOLTZ,AIDEN', 'SERAPH,REGINALD JEAN', 'LATIFF,WAZIR', 'SHEPHARD,GRANT']

def track_player_substitutions(data_frame, starting_lineup):
    player_substitutions = []
    current_lineup = starting_lineup.copy()
    lineup_history = []

    for _, row in data_frame.iterrows():
        action = row['First_True_Column']
        player = row['Second_True_Column']

        if action == 'enters the game':
            if len(current_lineup) >= 5:
                player_out = current_lineup.pop(0)
                current_lineup.append(player)
                player_substitutions.append((player_out, player))
        elif action == 'goes to the bench':
            if player in current_lineup:
                current_lineup.remove(player)
                if len(current_lineup) < 5:
                    player_in = current_lineup[-1]
                    current_lineup.append(player_in)
                    player_substitutions.append((player, player_in))

        lineup_history.append(current_lineup.copy())

    return player_substitutions, lineup_history

player_substitutions, lineup_history = track_player_substitutions(df, starter)
print(player_substitutions)
print(lineup_history)
```
