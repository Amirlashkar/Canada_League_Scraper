```python
# Extracting all unique players from the substitution columns
unique_players = pd.unique(filtered_rows1[['Second_True_Column']].values.ravel('K'))

# Initializing court times for all players
player_court_times_all = {player: 0 for player in unique_players if isinstance(player, str)}

# Reinitializing on-court status for all players (starting with everyone off-court)
on_court_all = {player: False for player in unique_players if isinstance(player, str)}

# Resetting last timestamp for the start of the quarter
last_timestamp_q1 = filtered_rows1['Time_sec'].iloc[-1]

# Iterate through each substitution event for all players
for index, row in filtered_rows1.iterrows():
    current_time_stamp = row['Time_sec']
    player = row['Second_True_Column']

    if player in on_court_all:
        # Update court time for players who are currently on the court
        for p in on_court_all:
            if on_court_all[p]:
                player_court_times_all[p] += last_timestamp_q1 - current_time_stamp

        # Update on-court status
        if row['First_True_Column'] == 'enters the game':
            on_court_all[player] = True
        elif row['First_True_Column'] == 'goes to the bench':
            on_court_all[player] = False

    # Update the last timestamp
    last_timestamp_q1 = current_time_stamp

# Add remaining time for players who were on the court at the end of the quarter
for p in on_court_all:
    if on_court_all[p]:
        player_court_times_all[p] += last_timestamp_q1

# Convert court times from seconds to minutes for all players
player_court_times_all_minutes = {player: time / 60 for player, time in player_court_times_all.items()}
player_court_times_all_minutes
```

```python
def calculate_player_points(data, scoring_values):
    """
    Calculate the total points scored by each player in a basketball game.

    Parameters:
    data (pd.DataFrame): DataFrame containing detailed play-by-play data.
    scoring_values (dict): Dictionary mapping scoring actions to their point values.

    Returns:
    dict: Total points scored by each player.
    """
    # Extracting player columns
    player_columns = data.columns[8:-2]

    # Initialize a dictionary to store the total points scored by each player
    player_points = {player: 0 for player in player_columns}

    # Function to calculate points based on scoring actions with the provided values
    def calculate_points(row):
        points = 0
        for action, value in scoring_values.items():
            if row[action]:
                points += value
        return points

    # Calculate points for each event in the game
    data['Points Scored'] = data.apply(calculate_points, axis=1)

    # Update points for each player
    for index, row in data.iterrows():
        for player in player_columns:
            if row[player]:
                player_points[player] += row['Points Scored']

    # Filter out players who didn't score any points
    player_points = {player: points for player, points in player_points.items() if points > 0}
    return player_points
```

```python
# Function to calculate points conceded by each player while on the court

def calculate_player_points_conceded_home(data, player_court_times):
    """
    Calculate the total points conceded by each player while they were on the court based on 'Home score'.

    Parameters:
    data (pd.DataFrame): DataFrame containing detailed play-by-play data.
    player_court_times (dict): Dictionary with players and their time on the court in minutes.

    Returns:
    dict: Total points conceded by each player while on the court.
    """
    # Initialize a dictionary to store points conceded while each player was on the court
    player_points_conceded = {player: 0 for player in player_court_times.keys()}

    # Calculate the difference in 'Home score' to determine points conceded in each event
    data['Points Conceded'] = data['Home score'].diff().fillna(0)

    # Iterate through the dataset and update points conceded for each player based on their court time
    for index, row in data.iterrows():
        for player in player_court_times.keys():
            if player_court_times.get(player, 0) > 0:  # Ensuring player is in the dictionary
                # Conceded points are counted for the visitor team's actions against the home team
                if 'Visitor' in row and row['Visitor'] and player in data.columns[28:]:  # Assuming Visitor players are in these columns
                    player_points_conceded[player] += row['Points Conceded']

    # Filter out players who didn't concede any points
    player_points_conceded = {player: points for player, points in player_points_conceded.items() if points > 0}
    return player_points_conceded

# Calculate points conceded for each player


```
